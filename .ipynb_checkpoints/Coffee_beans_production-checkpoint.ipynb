# Introduction
In this notebook, coffee beans production data is beeing explored. The data is taken from Food and Agriculture Organization of the United Nations (FAO) - https://www.fao.org/faostat/en/#data/QCL. The data inludes coffee beans (green) production data by country between 1961 and 2021. Also it includes information about harvested area and yield.

# Background
A coffee bean is a seed of Coffea and it is used to make coffee drink. As coffee drink is one of the world's most widely concumed beverage, coffee beans are very important product.

## Data investigation and preparation

<b>Let's take a brief look at the data.<b>

```python
df=pd.read_csv("FAOSTAT_data_en_6-16-2023.csv")
df.head()
```

```python
df.info()
```

<b>There are no NULL values.<b>


```python
df.describe()
```

<b>Column "Value" contains 0. Further investigation is required.<b>

<b>Let's investigate column by column (or group of columns).<b>

```python
df[["Domain Code","Domain"]].value_counts()
```

<b>Columns "Domain code", "Domain" do not contain any important information. It can be deleted.<b>

```python
df["Area"].unique()
```

<b>...'China', 'China, mainland', 'China, Taiwan Province of'...!? Further investigation is required.<b>

```python
df[df["Area"].str.contains("China")][["Area","Area Code (M49)"]].value_counts()
```

```python
df_china=df[df["Area Code (M49)"].isin([156,159,158])]
df_china[df_china["Year"]==1969]
```

```python
df_china_pivot=df_china.pivot(index=["Year","Element"],columns="Area",values="Value").reset_index()
df_china_pivot
```

```python
df_china_pivot[df_china_pivot.isna().any(axis=1)]
```

```python
df_china_pivot[df_china_pivot["China"].isna()]
```

```python
df_china_pivot[df_china_pivot["China, Taiwan Province of"].isna()]
```

```python
df_china_pivot[df_china_pivot["China, mainland"].isna()]
```

<b>Records related to "China" must be deleted, because it duplicates information: <b>
    
- Production and Area harvested for "China" is sum of "China, Taiwan Province of" and "China, mainland";
   

```python
df=df[df["Area Code (M49)"]!=156]
```

```python
df[["Item Code (CPC)","Item"]].value_counts()
```

<b> Columns "Item Code (CPC)", "Item" do not contain any important information. It can be deleted. <b>

```python
df[df["Year Code"]!=df["Year"]]
```

<b> "Year Code" duplicates "Year". Column "Year Code" can be deleted"<b>.

```python
df[["Flag Description","Flag"]].value_counts()
```

<b> Column "Flag Description" duplicates column "Flag". Column "Flag Description" can be deleted. <b>

```python
df.groupby("Flag")["Value"].describe()
```

<b>Column "Flag" also can be deleted, because it will not be used at this moment. <b> 

```python
df[["Element","Unit"]].value_counts()
```

<b> Column "Unit" can be deleted <b>

<b>Let's delete (drop) all unnecessary columns at once.<b>

```python
df.drop(["Domain Code","Domain","Item Code (CPC)","Item","Element Code","Year Code","Flag Description", "Flag", "Unit"],axis=1,inplace=True)
```

```python
df.head()
```

## Total coffee beans production by year

```python
plt.figure(figsize=(14,6))
plt.grid()
plt.title("Coffee beans production")
ax=sns.lineplot(x="Year",y="Value", data=df[df["Element"]=="Production"].groupby("Year").sum(),marker="o")
```

## Total area harvested for Coffea by year

```python
plt.figure(figsize=(14,6))
plt.grid()
plt.title("Area harvested for Coffea")
ax=sns.lineplot(x="Year",y="Value", data=df[df["Element"]=="Area harvested"].groupby("Year").sum(),marker="o")
```

## TOP 5 countries in 2021

```python
df[(df["Year"]==2021) & (df["Element"]=="Production")].sort_values(by="Value",ascending=False)[["Area","Value"]].head(5)
```

<b> As it was expected, the biggest producer is Brazil. <b>

```python
df[(df["Year"]==2021) & (df["Element"]=="Area harvested")].sort_values(by="Value",ascending=False)[["Area","Value"]].head(5)
```

<b> As it was expected, the biggest area for coffea is in Brazil. Viet Nam is in the second place by production of coffee beans, but it doesn't fall into TOP5 by Area harvested. <b>
